## **4. Trabajando con Datos en Notebooks**
- ### **Importacion de bibliotecas: NumPy,Pandas, matplotlib (Arriba, primera sección).**


- ###  **Carga y manipulacion basica de datos.**

```python
# Supongamos que estas son las mediciones de presión del agua durante una semana, en psi
presiones = np.array([30, 32, 31, 29, 30, 31, 33])

# Convertimos el array de NumPy en un DataFrame de Pandas
df_presiones = pd.DataFrame(presiones, columns=['Presión'])

# Añadimos una columna con los días de la semana
df_presiones['Día'] = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']

# Reordenamos las columnas
df_presiones = df_presiones[['Día', 'Presión']]

# Mostramos el DataFrame
print(df_presiones)

# Mostrar la información de las columnas del DataFrame
print(df_presiones.info())
```

- ### **Visualizacion con Matplotlib.**



```python
import matplotlib.pyplot as plt

# Asumiendo que ya tienes el DataFrame df_presiones como se definió anteriormente

# Graficamos las presiones a lo largo de la semana
plt.figure(figsize=(10, 5)) # Configuramos el tamaño del gráfico
plt.plot(df_presiones['Día'], df_presiones['Presión'], marker='o') # Creamos el gráfico de líneas

# Añadimos títulos y etiquetas
plt.title('Presión del Agua Durante la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Presión (psi)')

# Añadimos una cuadrícula para mejor visualización
plt.grid(True)

# Mostramos el gráfico
plt.show()

```

## **5. Ejemplo de Analisis de Datos**
- ### **Realización de un análisis simple utilizando los datos cargados.**


- ### **Generacion de graficos descriptivos.**

```python
import seaborn as sns

# Cargamos el dataset
df = pd.read_csv('2022-11-21 World Cup teams by country past week.csv')

# Mostramos las primeras filas del dataframe para verificar su contenido
print(df.head())

# Contamos la frecuencia con la que cada equipo aparece como 'Top searched'
conteo_equipos = df['Top searched'].value_counts()

# Gráfico de Barras de Frecuencia
plt.figure(figsize=(10, 5))
conteo_equipos.plot(kind='bar', color='skyblue')
plt.title('Frecuencia de Equipos Más Buscados')
plt.xlabel('Equipo')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Creamos una tabla de contingencia para el mapa de calor
tabla_contingencia = pd.crosstab(df['Country'], df['Top searched'])

# Mapa de Calor
plt.figure(figsize=(12, 8))
sns.heatmap(tabla_contingencia, annot=False, cmap='coolwarm')
plt.title('Mapa de Calor de Búsquedas por País y Equipo')
plt.xlabel('Equipo')
plt.ylabel('País')
plt.show()

```

------

 ### **Interpretacion de resultados**


#### ***Diagrama de Frecuencia***

*El diagrama de barras de frecuencia muestra cuántas veces cada equipo aparece como el más buscado en el conjunto de datos. Un equipo con una barra más alta indica que ha generado más interés de búsqueda en comparación con los demás. Esto puede reflejar su popularidad o relevancia durante el período de tiempo específico del dataset. Por ejemplo, si un equipo tiene una barra significativamente más alta que los demás, podría sugerir que tuvo un desempeño destacado en la Copa Mundial o que estuvo involucrado en eventos que captaron la atención del público.*

#### ***Mapa de Calor***

*El mapa de calor proporciona una representación visual de la relación entre dos variables categóricas: los países y los equipos más buscados. En este caso, los colores más cálidos (como el rojo) indican una mayor frecuencia de búsquedas de un equipo específico en un país, mientras que los colores más fríos (como el azul) indican una menor frecuencia. Este gráfico puede ayudar a identificar patrones geográficos en el interés de búsqueda, como qué equipos son populares en qué regiones o si hay algún equipo que destaque globalmente en términos de interés de búsqueda.*

*Ambos gráficos juntos pueden ofrecer una visión comprensiva de la dinámica de interés durante la Copa Mundial, mostrando tanto la popularidad general de los equipos como su distribución geográfica de interés.*
